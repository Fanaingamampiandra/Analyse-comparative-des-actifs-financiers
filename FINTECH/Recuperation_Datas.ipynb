```python

import yfinance as yf
import pandas as pd

# Définition des actifs
assets = {
    "Bitcoin": "BTC-USD",
    "Gold": "GC=F",
    "S&P500": "^GSPC"
}

# Définition de la période
start_date = "2019-01-01"
end_date = "2024-12-31"

# Récupération des données
data = {}
for asset, ticker in assets.items():
    df = yf.download(ticker, start=start_date, end=end_date)
    df["Asset"] = asset
    data[asset] = df

# Création d'un dataframe combiné
df_all = pd.concat(data.values())

# Affichage des premières lignes
df_all.head()
```

# Calcul de la Volatilité

```python
import numpy as np

# Calcul de la volatilité journalière
volatility_daily = df_all[["Return_BTC", "Return_Gold", "Return_SP500"]].std()

# Calcul de la volatilité annualisée

volatility_annualized = volatility_daily * np.sqrt(252)

# Affichage des résultats
print("Volatilité journalière des actifs :\n", volatility_daily)
print("\nVolatilité annualisée des actifs :\n", volatility_annualized)
```

```python
# Fenêtre mobile de 30 jours pour lisser la volatilité
df_all["Volatility_BTC"] = df_all["Return_BTC"].rolling(window=30).std() * np.sqrt(252)
df_all["Volatility_Gold"] = df_all["Return_Gold"].rolling(window=30).std() * np.sqrt(252)
df_all["Volatility_SP500"] = df_all["Return_SP500"].rolling(window=30).std() * np.sqrt(252)

# Tracé des volatilités
plt.figure(figsize=(12, 6))
plt.plot(df_all["Date"], df_all["Volatility_BTC"], label="Volatilité Bitcoin", color="orange", alpha=0.7)
plt.plot(df_all["Date"], df_all["Volatility_Gold"], label="Volatilité Gold", color="gold", alpha=0.7)
plt.plot(df_all["Date"], df_all["Volatility_SP500"], label="Volatilité S&P500", color="blue", alpha=0.7)

plt.xlabel("Date")
plt.ylabel("Volatilité Annualisée")
plt.title("Volatilité annualisée (Fenêtre mobile 30 jours)")
plt.legend()
plt.show()
```

# # TAUX DE CROISSANCE

```python
# Récupération des prix de 2019 et 2024
start_prices = df_all[df_all["Date"] == df_all["Date"].min()][["Close_BTC", "Close_Gold", "Close"]]
end_prices = df_all[df_all["Date"] == df_all["Date"].max()][["Close_BTC", "Close_Gold", "Close"]]

# Extraction des valeurs
start_prices = start_prices.iloc[0]
end_prices = end_prices.iloc[0]

# Nombre d'années (2019-2024)
n = 5

# Calcul du TCAC pour chaque actif
cagr_btc = ((end_prices["Close_BTC"] / start_prices["Close_BTC"]) ** (1/n)) - 1
cagr_gold = ((end_prices["Close_Gold"] / start_prices["Close_Gold"]) ** (1/n)) - 1
cagr_sp500 = ((end_prices["Close"] / start_prices["Close"]) ** (1/n)) - 1

# Stockage des résultats dans un dictionnaire
cagr_values = {
    "Bitcoin": cagr_btc * 100,
    "Gold": cagr_gold * 100,
    "S&P 500": cagr_sp500 * 100
}

# Affichage des résultats
print("Taux de Croissance Annuel Composé (CAGR) sur 5 ans :")
for asset, cagr in cagr_values.items():
    print(f"{asset} : {cagr:.2f} %")
```

```python
import matplotlib.pyplot as plt

# Correction : Convertir les valeurs du dictionnaire en liste
cagr_labels = list(cagr_values.keys())  # Noms des actifs
cagr_values_list = list(cagr_values.values())  # Valeurs du CAGR

# Création du graphique en barres
plt.figure(figsize=(8, 5))
plt.bar(cagr_labels, cagr_values_list, color=['orange', 'gold', 'blue'])
plt.xlabel("Actifs")
plt.ylabel("TCAC (%)")
plt.title("Taux de Croissance Annuel Composé (2019-2024)")
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Ligne de référence à 0%
plt.show()

```

# Évaluation des Risques
#Maximum Drawdown (MDD)

```python
# Fonction pour calculer le Maximum Drawdown
def calculate_mdd(price_series):
    rolling_max = price_series.cummax()  # Plus haut prix atteint jusqu'à chaque date
    drawdown = (price_series - rolling_max) / rolling_max  # Calcul des baisses
    max_drawdown = drawdown.min() * 100  # Le pire drawdown en pourcentage
    return max_drawdown

# Calcul du Maximum Drawdown pour chaque actif
mdd_btc = calculate_mdd(df_all["Close_BTC"])
mdd_gold = calculate_mdd(df_all["Close_Gold"])
mdd_sp500 = calculate_mdd(df_all["Close"])

# Affichage des résultats
mdd_values = {
    "Bitcoin": mdd_btc,
    "Gold": mdd_gold,
    "S&P 500": mdd_sp500
}

print("Maximum Drawdown des actifs (%) :")
for asset, mdd in mdd_values.items():
    print(f"{asset} : {mdd:.2f} %")
```

```python
import matplotlib.pyplot as plt

# Calcul des drawdowns sur le temps
df_all["Drawdown_BTC"] = (df_all["Close_BTC"] / df_all["Close_BTC"].cummax()) - 1
df_all["Drawdown_Gold"] = (df_all["Close_Gold"] / df_all["Close_Gold"].cummax()) - 1
df_all["Drawdown_SP500"] = (df_all["Close"] / df_all["Close"].cummax()) - 1

# Tracer les drawdowns
plt.figure(figsize=(12, 6))
plt.plot(df_all["Date"], df_all["Drawdown_BTC"], label="Bitcoin", color="orange", alpha=0.7)
plt.plot(df_all["Date"], df_all["Drawdown_Gold"], label="Gold", color="gold", alpha=0.7)
plt.plot(df_all["Date"], df_all["Drawdown_SP500"], label="S&P 500", color="blue", alpha=0.7)

plt.xlabel("Date")
plt.ylabel("Drawdown (%)")
plt.title("Évolution du Maximum Drawdown")
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)  # Ligne de référence
plt.legend()
plt.show()
```

# Calcul du Ratio de Sharpe

```python
# Définition du taux sans risque (par exemple, 2%)
risk_free_rate = 0.02

# Calcul du rendement moyen annuelisé (moyenne des rendements quotidiens x 252 jours)
mean_return_annualized = df_all[["Return_BTC", "Return_Gold", "Return_SP500"]].mean() * 252

# Utilisation des volatilités annualisées calculées précédemment
sharpe_ratios = (mean_return_annualized - risk_free_rate) / volatility_annualized

# Affichage des résultats
print("Rendement moyen annualisé des actifs :\n", mean_return_annualized)
print("\nVolatilité annualisée des actifs :\n", volatility_annualized)
print("\nRatio de Sharpe des actifs :\n", sharpe_ratios)
```

```python
import matplotlib.pyplot as plt

# Création du graphique
plt.figure(figsize=(8,5))
plt.bar(sharpe_ratios.index, sharpe_ratios.values, color=['orange', 'gold', 'blue'])
plt.xlabel("Actifs")
plt.ylabel("Ratio de Sharpe")
plt.title("Comparaison du Ratio de Sharpe pour Bitcoin, Gold et S&P 500")
plt.axhline(y=1, color='red', linestyle='--', label="Seuil de 1")
plt.legend()
plt.show()
```

# Calcul de la volatilité sur le mois de Janvier 2025  

```python
# Filtrer les données pour janvier 2025
df_jan_2025 = df_all[(df_all['Date'] >= '2025-01-01') & (df_all['Date'] <= '2025-01-31')]
```

```python
print(df_jan_2025.columns)
```

```python
# Renommer la colonne 'Close' en 'Close_SP500'
df_jan_2025.rename(columns={'Close': 'Close_SP500'}, inplace=True)
```

```python
import numpy as np

# Calcul des rendements journaliers logarithmiques
df_jan_2025['Return_BTC'] = np.log(df_jan_2025['Close_BTC'] / df_jan_2025['Close_BTC'].shift(1))
df_jan_2025['Return_Gold'] = np.log(df_jan_2025['Close_Gold'] / df_jan_2025['Close_Gold'].shift(1))
df_jan_2025['Return_SP500'] = np.log(df_jan_2025['Close_SP500'] / df_jan_2025['Close_SP500'].shift(1))
```

```python
import numpy as np

# Calcul des rendements journaliers logarithmiques en utilisant .loc[]
df_jan_2025.loc[:, 'Return_BTC'] = np.log(df_jan_2025['Close_BTC'] / df_jan_2025['Close_BTC'].shift(1))
df_jan_2025.loc[:, 'Return_Gold'] = np.log(df_jan_2025['Close_Gold'] / df_jan_2025['Close_Gold'].shift(1))
df_jan_2025.loc[:, 'Return_SP500'] = np.log(df_jan_2025['Close_SP500'] / df_jan_2025['Close_SP500'].shift(1))
```

```python
Volatilité Journalière
```

```python
# Calcul de la volatilité journalière (écart-type des rendements journaliers)
volatility_jan_2025 = {
    'Bitcoin': df_jan_2025['Return_BTC'].std(),
    'Gold': df_jan_2025['Return_Gold'].std(),
    'S&P 500': df_jan_2025['Return_SP500'].std()
}

# Affichage des volatilités journalières
for asset, vol in volatility_jan_2025.items():
    print(f"Volatilité journalière de {asset} en janvier 2025 : {vol:.4f}")
```

```python
Volatilité Annuelle
```

```python
# Nombre de jours de trading en janvier 2025
trading_days_jan_2025 = df_jan_2025['Return_BTC'].count()

# Facteur d'annualisation
annualization_factor = np.sqrt(252 / trading_days_jan_2025)

# Calcul de la volatilité annualisée
volatility_annualized_2025 = {asset: vol * annualization_factor for asset, vol in volatility_jan_2025.items()}

# Affichage des volatilités annualisées
for asset, vol in volatility_annualized_2025.items():
    print(f"Volatilité annualisée de {asset} pour 2025 : {vol:.4f}")
```

```python
import matplotlib.pyplot as plt

# Noms des actifs
assets = list(volatility_jan_2025.keys())

# Volatilité journalière en pourcentage
daily_volatility_percent = [vol * 100 for vol in volatility_jan_2025.values()]

# Volatilité annualisée en pourcentage
annualized_volatility_percent = [vol * 100 for vol in volatility_annualized_2025.values()]

# Création de la figure et des sous-graphiques
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Graphique de la volatilité journalière
ax[0].bar(assets, daily_volatility_percent, color=['orange', 'gold', 'blue'])
ax[0].set_title('Volatilité Journalière Moyenne en Janvier 2025')
ax[0].set_xlabel('Actifs')
ax[0].set_ylabel('Volatilité (%)')

# Graphique de la volatilité annualisée
ax[1].bar(assets, annualized_volatility_percent, color=['orange', 'gold', 'blue'])
ax[1].set_title('Volatilité Annualisée pour l\'Année 2025')
ax[1].set_xlabel('Actifs')
ax[1].set_ylabel('Volatilité (%)')

# Affichage des graphiques
plt.tight_layout()
plt.show()
```
